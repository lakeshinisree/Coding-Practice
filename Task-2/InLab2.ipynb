```python
# program to calculate the total amount of rainwater that can be trapped between buildings represented by an array of heights
def trap_rainwater(heights):
    if len(heights) < 3:
        return 0

    left = 0
    right = len(heights) - 1
    left_max = 0
    right_max = 0
    water = 0

    while left < right:
        if heights[left] < heights[right]:
            if heights[left] >= left_max:
                left_max = heights[left]
            else:
                water += left_max - heights[left]
            left += 1
        else:
            if heights[right] >= right_max:
                right_max = heights[right]
            else:
                water += right_max - heights[right]
            right -= 1

    return water


# User Input
n = int(input("Enter the number of buildings: "))
heights = list(map(int, input("Enter the heights separated by space: ").split()))

if len(heights) != n:
    print("Error: Number of heights does not match.")
else:
    print("Total Rainwater Trapped =", trap_rainwater(heights), "units")
```

```python
# Program to determine the maximum possible sum of a circular sub array.
def kadane(arr):
    max_sum = arr[0]
    current_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)

    return max_sum


def max_circular_subarray_sum(arr):
    # Maximum subarray sum (normal)
    max_kadane = kadane(arr)

    # If all elements are negative
    if max_kadane < 0:
        return max_kadane

    # Find total sum and invert array
    total_sum = sum(arr)
    inverted = [-x for x in arr]

    # Maximum circular sum
    max_wrap = total_sum + kadane(inverted)

    return max(max_kadane, max_wrap)


# User Input
n = int(input("Enter the number of elements: "))
arr = list(map(int, input("Enter the elements separated by space: ").split()))

if len(arr) != n:
    print("Error: Number of elements does not match.")
else:
    print("Maximum Circular Subarray Sum =", max_circular_subarray_sum(arr))
```

```python
# Program to determine the maximum profit that can be earned by buying and selling a stock exactly once using the given daily stock prices.
def max_profit(prices):
    if len(prices) < 2:
        return 0

    min_price = prices[0]
    max_profit = 0

    for price in prices[1:]:
        if price < min_price:
            min_price = price
        else:
            profit = price - min_price
            if profit > max_profit:
                max_profit = profit

    return max_profit


# User Input
n = int(input("Enter the number of days: "))
prices = list(map(int, input("Enter the stock prices separated by space: ").split()))

if len(prices) != n:
    print("Error: Number of prices does not match.")
else:
    print("Maximum Profit =", max_profit(prices))
```

```python
# rogram to determine the minimum number of candies required to distribute among children based on their ratings while satisfying the following constraints:
# • Every child must receive at least one candy.
# • A child with a higher rating than an immediate neighbour must receive more candies than that neighbour.
# • Determine the minimum number of candies needed.
def minimum_candies(ratings):
    n = len(ratings)

    # Initially, give each child 1 candy
    candies = [1] * n

    # Left to Right
    for i in range(1, n):
        if ratings[i] > ratings[i - 1]:
            candies[i] = candies[i - 1] + 1

    # Right to Left
    for i in range(n - 2, -1, -1):
        if ratings[i] > ratings[i + 1]:
            candies[i] = max(candies[i], candies[i + 1] + 1)

    return sum(candies), candies


# User Input
n = int(input("Enter the number of children: "))
ratings = list(map(int, input("Enter the ratings separated by space: ").split()))

if len(ratings) != n:
    print("Error: Number of ratings does not match.")
else:
    total, candies = minimum_candies(ratings)

    print("\nCandy Distribution:")
    print("Child     :", end=" ")
    for i in range(1, n + 1):
        print(i, end=" ")
    print()

    print("Rating    :", end=" ")
    for r in ratings:
        print(r, end=" ")
    print()

    print("Candies   :", end=" ")
    for c in candies:
        print(c, end=" ")
    print()

    print("\nMinimum Candies Required =", total)
```

```python
# Program to determine whether two strings are isomorphic.
def is_isomorphic(s1, s2):
    if len(s1) != len(s2):
        return False

    map1 = {}
    map2 = {}

    for i in range(len(s1)):
        ch1 = s1[i]
        ch2 = s2[i]

        if ch1 in map1:
            if map1[ch1] != ch2:
                return False
        else:
            map1[ch1] = ch2

        if ch2 in map2:
            if map2[ch2] != ch1:
                return False
        else:
            map2[ch2] = ch1

    return True


# User Input
str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")

if is_isomorphic(str1, str2):
    print("The strings are Isomorphic.")
else:
    print("The strings are Not Isomorphic.")
```

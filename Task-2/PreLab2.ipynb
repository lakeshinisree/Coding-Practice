```python
# Program to find the equilibrium index of an array.
# Equilibrium Index

arr = list(map(int, input("Enter array elements: ").split()))

for i in range(len(arr)):
    left = sum(arr[:i])
    right = sum(arr[i+1:])

    if left == right:
        print("Equilibrium Index:", i)
        break
else:
    print("No Equilibrium Index")
```

```python
# Program to rearrange an array such that all negative numbers appear before positive numbers while preserving their relative order.
# Rearrange array: Negative numbers first, then positive numbers

arr = list(map(int, input("Enter array elements: ").split()))

negative = []
positive = []

for num in arr:
    if num < 0:
        negative.append(num)
    else:
        positive.append(num)

result = negative + positive

print("Rearranged Array:", *result)
```

```python
#  Program to determine whether a given integer is a happy number
#Happy Number

num = int(input("Enter a number: "))

while num != 1 and num != 4:
    sum = 0
    while num > 0:
        digit = num % 10
        sum = sum + digit * digit
        num = num // 10
    num = sum

if num == 1:
    print("Happy Number")
else:
    print("Not a Happy Number")
```

```python
# Program to find all pairs of elements in an array whose absolute difference is equal to a given value.
# Find pairs with given absolute difference

arr = list(map(int, input("Enter array elements: ").split()))
diff = int(input("Enter difference: "))

found = False

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if abs(arr[i] - arr[j]) == diff:
            print(arr[i], arr[j])
            found = True

if not found:
    print("No pairs found")
```

```python
# Program to compute the frequency of each element in an array without using additional arrays.
# Frequency of each element in an array

arr = list(map(int, input("Enter array elements: ").split()))

for i in range(len(arr)):
    count = 1

    if arr[i] == -1:
        continue

    for j in range(i + 1, len(arr)):
        if arr[i] == arr[j]:
            count += 1
            arr[j] = -1

    print(arr[i], "->", count)
```

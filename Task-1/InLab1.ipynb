```python
#Program to solve Two Sum Problem
def two_sum(arr, target):
    d = {}

    for i in range(len(arr)):
        x = target - arr[i]

        if x in d:
            print("Indices:", d[x], "and", i)
            print("Elements:", arr[d[x]], "and", arr[i])
            return

        d[arr[i]] = i

    print("-1 (No valid pair found)")


# User Input
arr = list(map(int, input("Enter array elements: ").split()))
target = int(input("Enter target: "))

two_sum(arr, target)
```

```python
#Program to find the Longest Increasing Subsequence (LIS) of an array
def lis(arr):
    n = len(arr)
    dp = [1] * n

    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j]:
                dp[i] = max(dp[i], dp[j] + 1)

    print("Length of LIS:", max(dp))


# User Input
arr = list(map(int, input("Enter array elements: ").split()))

lis(arr)
```

```python
#Program to find the product of all elements in an array except the current element, without using the division operator.
def product_except_self(arr):
    n = len(arr)
    result = [1] * n

    # Left products
    left = 1
    for i in range(n):
        result[i] = left
        left *= arr[i]

    # Right products
    right = 1
    for i in range(n - 1, -1, -1):
        result[i] *= right
        right *= arr[i]

    print("Output:", *result)


# User Input
arr = list(map(int, input("Enter array elements: ").split()))

product_except_self(arr)
```

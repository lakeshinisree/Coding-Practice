```python
#Program to generate the Fibonacci series up to N terms.
# Input number of terms
n = int(input("Enter the number of terms: "))

first = 0
second = 1

print("Fibonacci Series:")

for i in range(n):
    print(first, end=" ")

    next = first + second
    first = second
    second = next
```

```python
#Program to find whether a given number is prime or not.
n = int(input("Enter a number: "))

count = 0

for i in range(1, n + 1):
    if n % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
#Program to find the majority element in an array using an efficient algorithm
# Input array
arr = list(map(int, input("Enter array elements: ").split()))

n = len(arr)
majority = -1

for i in arr:
    count = 0
    for j in arr:
        if i == j:
            count += 1

    if count > n // 2:
        majority = i
        break

if majority != -1:
    print("Majority Element:", majority)
else:
    print("No Majority Element")
```

```python
#Program to identify duplicate elements in an array
# Input array
arr = list(map(int, input("Enter array elements: ").split()))

print("Duplicate Elements:")

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] == arr[j]:
            print(arr[i])
            break
```

```python
#Program to identify most repetitive number present in the array.
# Input array
arr = list(map(int, input("Enter array elements: ").split()))

max_count = 0
most_repeated = arr[0]

for i in arr:
    count = 0
    for j in arr:
        if i == j:
            count += 1

    if count > max_count:
        max_count = count
        most_repeated = i

print("Most Repetitive Number:", most_repeated)
print("Repeated", max_count, "times")
```

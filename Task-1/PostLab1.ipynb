```python
#Program to find the next greater element for every element in an array
# Next Greater Element

arr = list(map(int, input("Enter array elements: ").split()))

for i in range(len(arr)):
    found = -1
    for j in range(i + 1, len(arr)):
        if arr[j] > arr[i]:
            found = arr[j]
            break
    print(arr[i], "->", found)
```

```python
#Program to find all triplets in an array whose sum is equal to a given target value (3Sum Problem)
# 3Sum Problem

arr = list(map(int, input("Enter array elements: ").split()))
target = int(input("Enter target: "))

n = len(arr)
found = False

for i in range(n):
    for j in range(i + 1, n):
        for k in range(j + 1, n):
            if arr[i] + arr[j] + arr[k] == target:
                print(arr[i], arr[j], arr[k])
                found = True

if not found:
    print("No triplets found")
```

```python
#Program to find the first missing positive integer in an unsorted array.
#First Missing Positive Integer

arr = list(map(int, input("Enter array elements: ").split()))

num = 1

while True:
    if num not in arr:
        print("First Missing Positive Integer:", num)
        break
    num += 1
```
